```python
import ijson
import numpy as np

paper_written_by_author=[["id_author", "id_paper"]]
paper_published_in_venue=[["id_paper","id_venue", "edition", "year", "raw"]]
paper_published_in_journal=[["id_paper","id_journal", "volume", "year", "raw"]]
paper_contains_keywords=[["id_paper", "id_keyword"]]
limit_data=10000
```

```python
!pip install tqdm
```

```python
import csv
import tqdm
import ijson
```

```python
with open("dblp.v12.json", "rb") as f:
    for i, paper in tqdm.tqdm(enumerate(ijson.items(f, "item"))):
        id_paper= paper["id"]
        authors = paper["authors"]
        for j in authors:
            paper_written_by_author.append([id_paper, j["id"]])
        #paper_published_in_venue.csv (edition, year, venue)
        venue_paper=paper["venue"]
        if("type" in venue_paper):
            if(venue_paper["type"]=="C" and len(paper_published_in_venue)<limit_data):
                paper_published_in_venue.append([id_paper,venue_paper["id"], paper["volume"],paper["year"] ])
            #paper_published_in_journal.csv (volume, year)
            elif(venue_paper["type"]=="J" and len(paper_published_in_journal)<limit_data):
                paper_published_in_journal.append([id_paper,venue_paper["id"], paper["volume"],paper["year"] ])
        
        #paper_contains_keywords.csv
        if("fos" in paper):
            for j in paper["fos"]:
                paper_contains_keywords.append([id_paper,j["name"]])
        if(len(paper_published_in_venue)>=limit_data and len(paper_published_in_journal)>=limit_data):
            break

```

```python
paper_written_by_author_t=paper_written_by_author[1:len(paper_written_by_author)]
```

```python
#Save files
arr = np.asarray(paper_written_by_author_t)
np.savetxt('./paper_written_by_author.csv',arr, delimiter=",", header='id_author,id_paper') 
```

```python
paper_published_in_venued=paper_published_in_venue[1:len(paper_published_in_venue)]
paper_published_in_journald=paper_published_in_journal[1:len(paper_published_in_journal)]
paper_contains_keywordsd=paper_contains_keywords[1:len(paper_contains_keywords)]
```

```python

arr = np.asarray(paper_published_in_venued)
np.savetxt('./paper_published_in_venue.csv',arr, fmt='%s', delimiter=",", header="id_paper,id_venue,edition,year,raw") 

arr = np.asarray(paper_published_in_journald)
np.savetxt('./paper_published_in_journal.csv',arr, fmt='%s', delimiter=",", header="id_paper,id_journal,volume,year,raw") 

arr = np.asarray(paper_contains_keywordsd)
np.savetxt('./paper_contains_keywords.csv',arr, fmt='%s', delimiter=",", header="id_paper,id_keyword") 
```
